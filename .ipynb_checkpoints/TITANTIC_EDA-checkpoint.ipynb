```python
import pandas as pd

# Update the path to match your file's location
data = pd.read_csv('/Users/rohandhurai/Documents/Titanic Data/archive/train.csv')

# Display the first 5 rows
print(data.head())
```

```python
# Check the column names, data types, and non-null counts
print(data.info())

# See basic statistics for numerical columns
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
data.drop('Cabin', axis=1, inplace=True)
```

```python
print(data.isnull().sum())
```

```python
print(data['Survived'].value_counts(normalize=True) * 100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=data, kind='bar')
plt.title('Survival Rate by Gender and Class')
plt.show()
```

```python
sns.histplot(data=data, x='Age', hue='Survived', kde=True, multiple='stack')
plt.title('Age Distribution: Survivors vs Non-Survivors')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Suppress the FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning)

# Replace infinite values with NaN
data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Plot survival rate by gender
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.histplot(data=data, x='Age', hue='Survived', kde=True, multiple='stack')
plt.title('Age Distribution: Survivors vs Non-Survivors')
plt.show()
```
